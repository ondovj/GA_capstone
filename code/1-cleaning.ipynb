### Duplicate Energy

When previously attempting to combine and pivot the nutrients so that they can be transferred into a final food dataframe, it was noticed that there were duplicate nutrient names preventing the transposition. This duplicate entry needs to be identified and dealt with before the data can be brought together.

```python
# finding duplicated nutrient name
nutrient_names[nutrient_names["name"].duplicated()]
```

There appears to be a second "energy" entry, which is in kilojoules, instead of kilocalories, which are typical for US foods. This will be further explored in the `nutrient_info`to see how many entries are in kJ.

```python
nutrient_info[nutrient_info["nutrient_id"] == 1062]["amount"]
```

We have nine foods using kJ as its energy, instead of kcal. This is an easy fix, as we can simply convert to the proper format. This is a common conversion, as energy is commonly represented as kilojoules outside of the US. By using a reference from the University of Calgary, we can find that 1 kilocalorie is equal to 4.1868 kilojoules<sup>(1)</sup>.

```python
# mapping the conversion 
nutrient_info.loc[nutrient_info["nutrient_id"] == 1062, "amount"] = nutrient_info[
    nutrient_info["nutrient_id"] == 1062]["amount"].map(lambda x: round(x / 4.1868, 2))
```

```python
# checking for the updated values
nutrient_info[nutrient_info["nutrient_id"] == 1062]
```

Now that the values have been converted, we can change the `nutrient_id` entries to be in line with the kilocalorie ID. For this, we have to look up the target ID from the `nutrient_names` dataframe.

```python
# checking the id of energies
nutrient_names[nutrient_names["name"] == "Energy"]
```

If we can now convert the ID of the kJ entries to be the KCAL ID, we can merge the two dataframes, and the names should sync up with the KCAL name instead.

```python
# checking the count of kJ entries before replacement
nutrient_info[nutrient_info["nutrient_id"] == 1062].count()
```

```python
# repalcing the id numbers
nutrient_info["nutrient_id"].replace(to_replace=1062, value=1008, inplace=True)
```

```python
# checking the count of kJ entries after replacement
nutrient_info[nutrient_info["nutrient_id"] == 1062].count()
```

```python
nutrient_info[nutrient_info["nutrient_id"] == 1008]["derivation_id"].value_counts()
```

### Nurtient Pivot Table

```python
# making pivot table of nutrients df
nutrients_pivot = nutrients.pivot_table(index = 'fdc_id', columns = 'name', values = 'amount')
nutrients_pivot.head()
```

The first thing to notice with this new dataframe is the large amount of null values. Since not every food has each nutrient listed on its label, this is not completely unexpected. However, we do need to find out the extent of the nulls in this new version of the dataset.

```python
nutrients_pivot.isnull().mean().sort_values()
```

It appears as though a large percent of these columns are mostly null. This can be explained by the fact that only a small set of nutrients are actually required to be listed on food product labels. By referencing the Code of Federal Regulations Title 21, which covers food for human consumption, we can learn a bit more about how nutrient labels are constructed<sup>(3)</sup>.

There are only a few nutrients that are required to be on nutrition labels:
 - Calories
 - Total fat
 - Cholesterol*
 - Sodium
 - Carbohydrate total
     - Dietary fiber*
     - Total sugars*
     - Added sugars*
 - Protein*
 
 Even within these nutrients, those marked with asterisks have additional caveats where certain foods are not required to list these on the nutrient label if there is no presence of the nutrient, and there are no other claims or declarations made about those nutrients on the packaging. So, many of these null values may actually indicate a value of `0`, but was recorded as null because it was not even present on the label. Taking this into account, we can see why the amount of null values increases sharply outside these core nutrients.
 
 In order to effectively analyze the foods we have, we are going to have to limit the amount of these nutrients we utilize as features. At the same time, we want to still include as much as possible, and there may be ways of estimating some of the missing values using other context clues from the foods. As vitamins and minerals are largely voluntary, they will be excluded (except for sodium). Outside of these, we will take the top 10 nutrients.

```python
# grabbing the top 12 rows
nutrients_pivot.isnull().mean().sort_values().head(12).index
```

```python
# making list based of top 10 non-vitamin/mineral
nutr_cols = ['Total lipid (fat)', 'Protein', 'Carbohydrate, by difference',
       'Sodium, Na', 'Energy', 'Sugars, total including NLEA',
       'Fatty acids, total saturated', 'Cholesterol', 'Fiber, total dietary',
            'Fatty acids, total trans']
```

```python
# selecting for desired cols
nutrients_pivot = nutrients_pivot[nutr_cols]

# checking new df
nutrients_pivot.head()
```

```python
# checking info on new df
nutrients_pivot.describe()
```

Now that we have narrowed down our nutrient columns, we seem to have actually taken care of one of the previous issues that was spotted, which were some suspiciously high values for the nutrient amounts. ### Calculating Energy

From our null count above, we saw nearly 2000 null values in the `energy` column, meaning the calorie count is missing. Fortunately, there is an easy way to calculate calories from the surrounding nutrient context. With some reserach, it can be found that the calories from any food come from three major macronutrient groups: fat, carbohydrates, and protein. Fats supply 9 calories/gram, while carbohydrates and protein both give 4 calories/gram<sup>(4)</sup>. Therefore, we can use a simple mathematical calculation to fill in the actual calories.

```python
# checking starting nulls for energy
foods["energy"].isnull().sum()
```

```python
# making new col with the calc
foods["null_energy"] = (foods["fat_total"] * 9) + (foods["carbs"] * 4) + (foods["protein"] * 4)

# checking for new col
foods.head()
```

We can see from the head of the dataframe that this calculated value isn't perfectly aligning with some of the entries that did already have a calorie count listed. This may be due the specific lab methods used to determine the calorie content, since these values are not typically calculated the way that we have done here

```python
# inserting new val into null energy spots
foods.loc[foods["energy"].isnull(), "energy"] = foods.loc[foods["energy"].isnull(), "null_energy"]
```

```python
# rechecking the sum of null energies
foods["energy"].isnull().sum()
```

```python
# removing the extra col
foods.drop(columns="null_energy", inplace=True)
```

We have now managed to get reduce our null energy values to below 15 entries, down from nearly 1800, just by using a simple calculation. The remaining values must have nulls in one or more of the macronutrient columns (fat, proetin, carbs). Some of the remaining entries may have many null values in their nutrients, potentially making them unreliable entries, and should most likely be dropped due to their lack of providing good information for the model.

### Cleaning Nutrients

We have a lot of entries in the saturated and trans fat columns that are null. Since there are many foods that have 0g fat, we can conclude that those foods would also have 0g of the sub-categories of fat.

This logic will also follow for sugars and fiber, which are subsets of carbohydrates.

```python
# setting saturated fats to 0 if total fat is 0
foods.loc[foods["fat_total"] == 0, "fat_sat"] = foods[foods["fat_total"] == 0]["fat_sat"].map(lambda x: 0)
```

```python
# repeating for trans fat
foods.loc[foods["fat_total"] == 0, "fat_trans"] = foods[foods["fat_total"] == 0]["fat_trans"].map(lambda x: 0)
```

```python
# setting fiber to 0 if total carbs is 0
foods.loc[foods["carbs"] == 0, "fiber"] = foods[foods["carbs"] == 0]["fiber"].map(lambda x: 0)
```

```python
# repeating for sugar
foods.loc[foods["carbs"] == 0, "sugars"] = foods[foods["carbs"] == 0]["sugars"].map(lambda x: 0)
```

```python
foods[(foods["fat_sat"].isnull()) & (foods["fat_trans"].isnull())].shape
```

```python
foods.shape
```

```python
foods.isnull().sum()
```

This has greatly reduced the number of null values across our nutrient columns. However, many null values remain for the four nutrients that were just dealt with. Since these columns can be voluntary, the manufacturers may not have the actual information of these specific nutrients. Since there is now no way for use to know what the values of these should be, we should remove them from our dataset. 

It is important to note that this may have unexpected impacts on the final analysis and modeling of this data, as this will be removing foods that may have informative values in these columns, but are not fully documented in this dataset. However, this step is being taken so that we can build a model using the most amount of nutritional information available. 

```python
# dropping remaining nulls for sugar, fiber, sat fat, and trans fat
foods.dropna(subset=["sugars", "fiber", "fat_sat", "fat_trans"], inplace=True)
```

The last major nutrient that has a large amount of null values is cholesterol. Like with some of the other values, foods are not necessarily required to list it on their label, if the value is zero or close to it, and it meets certain other labeling requirements. The difference with this column compared to some of the others that we have dealt with is that there really are no contextual clues as to the true value of this nutrient. Because of this, we will rely completely on the FDA guidelines, and are going to blanket impute the null values in this column with 0.

```python
# making null chol into 0
foods.loc[foods["chol"].isnull(), "chol"] = foods[foods["chol"].isnull()]["chol"].map(lambda x: 0)
```
